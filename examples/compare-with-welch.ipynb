```python
import numpy as np
import matplotlib.pyplot as plt
from pylpsd import lpsd
from scipy.signal import welch
```

```python

N = int(1e5)  # Number of data points in the timeseries
fs = 1.0
x = np.random.normal(size=N)


f, X = lpsd(x, fs, window='hann')

# Compare to Welch

# nperseg doesn't seem to be very appropriate but it was set like this in the old `pylpsd_demo.py`
f_welch, Pxx = welch(x, fs, window='hann', nperseg=N)

plt.figure()
plt.loglog(f_welch, Pxx)
plt.loglog(f, X)
```
